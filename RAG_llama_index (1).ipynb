```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, ServiceContext
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
documents=SimpleDirectoryReader("/content/data").load_data()
documents
```

```python
system_prompt="""
You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided.
"""
## Default format supportable by LLama2
query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True}
)
```

```python
!pip install -U langchain-community
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core.service_context import ServiceContext
from llama_index.legacy.embeddings.langchain import LangchainEmbedding

embed_model=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
service_context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
service_context
```

```python
index=VectorStoreIndex.from_documents(documents,service_context=service_context)
```

```python
index
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("what is attention is all you need?")
```

```python
# prompt: make this model into a voice assistant chatbot
!pip install speechrecognition
import speech_recognition as sr

# Initialize the speech recognition object
r = sr.Recognizer()

while True:
    # Prompt the user to speak
    print("Chatbot: Say something!")

    # Listen for the user's input
    with sr.Microphone() as source:
        audio = r.listen(source)

    try:
        # Convert the audio to text
        text = r.recognize_google(audio)

        # If the user said "exit", break out of the loop
        if text.lower() == "exit":
            break

        # Otherwise, query the model and print the response
        response = query_engine.query(text)
        print("Chatbot:", response)
    except sr.UnknownValueError:
        print("Chatbot: Sorry, I couldn't understand what you said.")
    except sr.RequestError as e:
        print("Chatbot: Error:", e)

# Say goodbye to the user
print("Chatbot: Goodbye!")
```

```python
# prompt: make this model into a voice assistant chatbot

# Import libraries for speech recognition and text-to-speech
!pip install speechrecognition
!pip install gtts

import speech_recognition as sr
import gtts

# Initialize the speech recognition and text-to-speech objects
r = sr.Recognizer()
# tts = gtts.gTTS()  <-- Remove this line. We will initialize gTTS when we need it in speak_text()

# Create a function to convert speech to text
def listen_to_speech():
  with sr.Microphone() as source:
    print("Listening...")
    audio_text = r.listen(source)
    print("Done listening.")
    try:
      # Convert speech to text
      recognized_text = r.recognize_google(audio_text)
      print("You: " + recognized_text)
      return recognized_text
    except sr.UnknownValueError:
      print("Speech Recognition could not understand audio.")
      return ""
    except sr.RequestError as e:
      print("Error occurred in connecting to the API. Check your internet connection. Error: ", e)
      return ""

# Create a function to convert text to speech
def speak_text(text):
  # Convert text to speech
  tts = gtts.gTTS(text=text, lang="en") # <-- Initialize gTTS here with the text to be spoken.
  # Save the audio file
  filename = "speech_output.mp3"
  tts.save(filename)
  # Play the audio file
  !mpg123 -q speech_output.mp3

# Start the chatbot loop
print("Chatbot: Hello! How can I help you today?")
while True:
  # Listen to speech input
  user_input = listen_to_speech()

  # Exit if user says "exit"
  if user_input.lower() == "exit":
    print("Chatbot: Goodbye!")
    break

  # Get response from the query engine
  response = query_engine.query(user_input)

  # Speak the response
  speak_text(response)
```
